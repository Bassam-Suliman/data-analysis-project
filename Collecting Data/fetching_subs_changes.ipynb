```python
from google.colab import files
import requests
import pandas as pd
```

```python
df = pd.read_csv('user2100.csv').copy()
handles = df.handle.to_list()
sum = 0
```

```python
for i in handles:
    try:
      content = requests.get(f'https://codeforces.com/api/user.status?handle={i}&from=1&count={20000}')
      submissions = pd.DataFrame(content.json()['result'])
      submissions.to_csv(f'/content/newf/{i}.csv', encoding='utf-8', index=False)
      sum += 1
      print(content.json()['status'],sum)
    except:
      print('Errored')
print('DONE !',sum)
```

```python
### Sumbissions are full of wrong answers, Let's clean
for i in handles:
    try:
        submissions = pd.read_csv(f'/content/newf/{i}.csv').copy()
        submissions = submissions[submissions.verdict == 'OK']
        submissions.drop(columns=['passedTestCount','timeConsumedMillis','memoryConsumedBytes','relativeTimeSeconds','creationTimeSeconds','id','programmingLanguage','points','testset','verdict','author'],inplace=True)
        submissions.drop_duplicates(inplace=True)
        def fun(row):
            #row = row['index']
            row = row.split()[3][1:-2]
            return row
        submissions['problem'] = submissions['problem'].apply(fun)
        submissions.to_csv(f'/content/final/{i}.csv', encoding='utf-8', index=False)
        pass
    except:
        continue
```

```python
contest = pd.read_csv('contests.csv').copy()
contest = contest.id.to_list()
sum = 0
```

```python
for i in contest:
  try:
    rate = pd.DataFrame(requests.get(f'https://codeforces.com/api/contest.ratingChanges?contestId={i}').json()['result'])
    rate = rate[rate['handle'].apply(lambda s: handles.__contains__(s))]
    rate.drop(columns=['contestName','ratingUpdateTimeSeconds'],inplace=True)
    rate.to_csv(f'/content/finalcon/{i}.csv', encoding='utf-8', index=False)
    sum += 1
    print('OK',sum)
  except:
    print('Errored')
```

```python
for i in contest:
  try:
    rate = pd.read_csv(f'/content/finalcon/{i}.csv').copy()
    rate.drop(columns=['contestId'],inplace=True)
    rate.to_csv(f'/content/morecon/{i}.csv', encoding='utf-8', index=False)
    sum += 1
    print('OK',sum)
  except:
    print('Errored')
```
