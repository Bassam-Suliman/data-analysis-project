```python
from bs4 import BeautifulSoup
import requests
```

```python
file = open('proxy.txt','w')
i = 1
while i < 300:
    proxa = "http://141.145.214.176:80"
    page = requests.get(f'https://hide.mn/en/proxy-list/?maxtime=200&start={i}#list', proxies={"http": proxa, "https": proxa})
    print(page)
    soup = BeautifulSoup(page.text, 'lxml')
    proxy_list = soup.find('div', class_='table_block')
    proxy_list = proxy_list.find_all('tr')
    for proxy in proxy_list:
        ip = proxy.find_all('td')[0].text
        port = proxy.find_all('td')[1].text
        file.write(f'{ip}:{port}')
    i+=70
file.close()
```
