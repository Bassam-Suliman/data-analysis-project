```python
### This python File adds number of solves for each problem
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from time import time
```

```python
prob = pd.read_csv('problems.csv',index_col=['contestId','index']).copy()
prob.dropna(axis=0,inplace=True)
prob['solved'] = 0
#prob.loc[(2035,'H'),'solved'] += 3
prob.head(10)
```

```python
users = pd.read_csv('user2100.csv',index_col='handle')
users.drop(columns=['maxRank','rank','friendOf','lstOnline'],inplace=True)
#user.sort_values(by='regTime',inplace=True)
users.head(5)
```

```python
A=0
B=0
for user in users.index:
    try:
        user_solved = pd.read_csv(f'Submissions/{user}.csv',index_col=['contestId','problem'])
        for row in user_solved.index:
            if row[0] > 3000:
                continue
            try:
                prob.loc[row,'solved'] += 1
            except:
                A += 1 
    except:
        B += 1
        print(user,'Failed!')
        pass
```

```python
print(A,B)
```
